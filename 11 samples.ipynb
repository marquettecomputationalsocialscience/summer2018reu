```python
ElevenMKE = Eleven [(Eleven['Longitude']>-88.080736) & (Eleven['Longitude']<-87.839722
)& (Eleven['Latitude']>42.91767) & (Eleven['Latitude']< 43.19712)]
# Create a boolean array setting boundaries to the map, Attribute these booleans to the columns of the original dataset


X = ElevenMKE['Longitude']
y = ElevenMKE['Latitude']


plt.clf()
plt.figure(figsize=(8, 8))
plt.scatter(X, y, s=60, alpha=.4, linewidths=0)
plt.title('Milwaukee Crime Events: Eleven [Cropped]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2011\Euclidean\Year\Milwaukee Crime Events Eleven [Cropped].png')
plt.show() 

ElevenKDE = pd.DataFrame(ElevenMKE, columns = ['Longitude','Latitude'])

KMfunction = KMeans(n_clusters = 5)
KMfunction.fit(ElevenKDE)
Colors = KMfunction.predict(ElevenKDE)

plt.clf()
plt.figure(figsize=(4, 4))
plt.scatter(ElevenKDE['Longitude'],ElevenKDE['Latitude'],s=60, c = Colors, cmap = 'tab10',alpha = .4, linewidths = 0)
plt.title('Milwaukee Crime Events: Eleven [K = 5]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2011\Euclidean\Year\Milwaukee Crime Events Eleven [K=5].png')
plt.show()

#******************************************************************************************************************************************************

KMfunction = KMeans(n_clusters = 4)
KMfunction.fit(ElevenKDE)
Colors = KMfunction.predict(ElevenKDE)

plt.clf()
plt.figure(figsize=(4, 4))
plt.scatter(ElevenKDE['Longitude'],ElevenKDE['Latitude'],s=60, c = Colors, cmap = 'tab10',alpha = .4, linewidths = 0)
plt.title('Milwaukee Crime Events: Eleven [K = 4]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2011\Euclidean\Year\Milwaukee Crime Events Eleven [K=4].png')
plt.show()

#******************************************************************************************************************************************************

KMfunction = KMeans(n_clusters = 3)
KMfunction.fit(ElevenKDE)
Colors = KMfunction.predict(ElevenKDE)

plt.clf()
plt.figure(figsize=(4, 4))
plt.scatter(ElevenKDE['Longitude'],ElevenKDE['Latitude'],s=60, c = Colors, cmap = 'tab10',alpha = .4, linewidths = 0)
plt.title('Milwaukee Crime Events: Eleven [K = 3]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2011\Euclidean\Year\Milwaukee Crime Events Eleven [K=3].png')
plt.show()

#******************************************************************************************************************************************************

KMfunction = KMeans(n_clusters = 6)
KMfunction.fit(ElevenKDE)
Colors = KMfunction.predict(ElevenKDE)

plt.clf()
plt.figure(figsize=(4, 4))
plt.scatter(ElevenKDE['Longitude'],ElevenKDE['Latitude'],s=60, c = Colors, cmap = 'tab10',alpha = .4, linewidths = 0)
plt.title('Milwaukee Crime Events: Eleven [K = 6]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2011\Euclidean\Year\Milwaukee Crime Events Eleven [K=6].png')
plt.show()

#******************************************************************************************************************************************************

KMfunction = KMeans(n_clusters = 7)
KMfunction.fit(ElevenKDE)
Colors = KMfunction.predict(ElevenKDE)

plt.clf()
plt.figure(figsize=(4, 4))
plt.scatter(ElevenKDE['Longitude'],ElevenKDE['Latitude'],s=60, c = Colors, cmap = 'tab10',alpha = .4, linewidths = 0)
plt.title('Milwaukee Crime Events: Eleven [K = 7]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2011\Euclidean\Year\Milwaukee Crime Events Eleven [K=7].png')
plt.show()

#******************************************************************************************************************************************************

```
