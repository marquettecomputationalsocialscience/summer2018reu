```python
TenMKE = Ten [(Ten['Longitude']>-88.080736) & (Ten['Longitude']<-87.839722
)& (Ten['Latitude']>42.91767) & (Ten['Latitude']< 43.19712)]
# Create a boolean array setting boundaries to the map, Attribute these booleans to the columns of the original dataset


X = TenMKE['Longitude']
y = TenMKE['Latitude']


plt.clf()
plt.figure(figsize=(8, 8))
plt.scatter(X, y, s=60, alpha=.4, linewidths=0)
plt.title('Milwaukee Crime Events: Ten [Cropped]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2010\Euclidean\Year\Milwaukee Crime Events Ten [Cropped].png')
plt.show() 

TenKDE = pd.DataFrame(TenMKE, columns = ['Longitude','Latitude'])

KMfunction = KMeans(n_clusters = 5)
KMfunction.fit(TenKDE)
Colors = KMfunction.predict(TenKDE)

plt.clf()
plt.figure(figsize=(4, 4))
plt.scatter(TenKDE['Longitude'],TenKDE['Latitude'],s=60, c = Colors, cmap = 'tab10',alpha = .4, linewidths = 0)
plt.title('Milwaukee Crime Events: Ten [K = 5]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2010\Euclidean\Year\Milwaukee Crime Events Ten [K=5].png')
plt.show()

#******************************************************************************************************************************************************

KMfunction = KMeans(n_clusters = 4)
KMfunction.fit(TenKDE)
Colors = KMfunction.predict(TenKDE)

plt.clf()
plt.figure(figsize=(4, 4))
plt.scatter(TenKDE['Longitude'],TenKDE['Latitude'],s=60, c = Colors, cmap = 'tab10',alpha = .4, linewidths = 0)
plt.title('Milwaukee Crime Events: Ten [K = 4]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2010\Euclidean\Year\Milwaukee Crime Events Ten [K=4].png')
plt.show()

#******************************************************************************************************************************************************

KMfunction = KMeans(n_clusters = 3)
KMfunction.fit(TenKDE)
Colors = KMfunction.predict(TenKDE)

plt.clf()
plt.figure(figsize=(4, 4))
plt.scatter(TenKDE['Longitude'],TenKDE['Latitude'],s=60, c = Colors, cmap = 'tab10',alpha = .4, linewidths = 0)
plt.title('Milwaukee Crime Events: Ten [K = 3]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2010\Euclidean\Year\Milwaukee Crime Events Ten [K=3].png')
plt.show()

#******************************************************************************************************************************************************

KMfunction = KMeans(n_clusters = 6)
KMfunction.fit(TenKDE)
Colors = KMfunction.predict(TenKDE)

plt.clf()
plt.figure(figsize=(4, 4))
plt.scatter(TenKDE['Longitude'],TenKDE['Latitude'],s=60, c = Colors, cmap = 'tab10',alpha = .4, linewidths = 0)
plt.title('Milwaukee Crime Events: Ten [K = 6]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2010\Euclidean\Year\Milwaukee Crime Events Ten [K=6].png')
plt.show()

#******************************************************************************************************************************************************

KMfunction = KMeans(n_clusters = 7)
KMfunction.fit(TenKDE)
Colors = KMfunction.predict(TenKDE)

plt.clf()
plt.figure(figsize=(4, 4))
plt.scatter(TenKDE['Longitude'],TenKDE['Latitude'],s=60, c = Colors, cmap = 'tab10',alpha = .4, linewidths = 0)
plt.title('Milwaukee Crime Events: Ten [K = 7]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2010\Euclidean\Year\Milwaukee Crime Events Ten [K=7].png')
plt.show()

#******************************************************************************************************************************************************

```
