```python
FourteenMKE = Fourteen [(Fourteen['Longitude']>-88.080736) & (Fourteen['Longitude']<-87.839722
)& (Fourteen['Latitude']>42.91767) & (Fourteen['Latitude']< 43.19712)]
# Create a boolean array setting boundaries to the map, Attribute these booleans to the columns of the original dataset


X = FourteenMKE['Longitude']
y = FourteenMKE['Latitude']


plt.clf()
plt.figure(figsize=(8, 8))
plt.scatter(X, y, s=60, alpha=.4, linewidths=0)
plt.title('Milwaukee Crime Events: Fourteen [Cropped]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2014\Euclidean\Year\Milwaukee Crime Events Fourteen [Cropped].png')
plt.show() 

FourteenKDE = pd.DataFrame(FourteenMKE, columns = ['Longitude','Latitude'])

KMfunction = KMeans(n_clusters = 5)
KMfunction.fit(FourteenKDE)
Colors = KMfunction.predict(FourteenKDE)

plt.clf()
plt.figure(figsize=(4, 4))
plt.scatter(FourteenKDE['Longitude'],FourteenKDE['Latitude'],s=60, c = Colors, cmap = 'tab10',alpha = .4, linewidths = 0)
plt.title('Milwaukee Crime Events: Fourteen [K = 5]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2014\Euclidean\Year\Milwaukee Crime Events Fourteen [K=5].png')
plt.show()

#******************************************************************************************************************************************************

KMfunction = KMeans(n_clusters = 4)
KMfunction.fit(FourteenKDE)
Colors = KMfunction.predict(FourteenKDE)

plt.clf()
plt.figure(figsize=(4, 4))
plt.scatter(FourteenKDE['Longitude'],FourteenKDE['Latitude'],s=60, c = Colors, cmap = 'tab10',alpha = .4, linewidths = 0)
plt.title('Milwaukee Crime Events: Fourteen [K = 4]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2014\Euclidean\Year\Milwaukee Crime Events Fourteen [K=4].png')
plt.show()

#******************************************************************************************************************************************************

KMfunction = KMeans(n_clusters = 3)
KMfunction.fit(FourteenKDE)
Colors = KMfunction.predict(FourteenKDE)

plt.clf()
plt.figure(figsize=(4, 4))
plt.scatter(FourteenKDE['Longitude'],FourteenKDE['Latitude'],s=60, c = Colors, cmap = 'tab10',alpha = .4, linewidths = 0)
plt.title('Milwaukee Crime Events: Fourteen [K = 3]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2014\Euclidean\Year\Milwaukee Crime Events Fourteen [K=3].png')
plt.show()

#******************************************************************************************************************************************************

KMfunction = KMeans(n_clusters = 6)
KMfunction.fit(FourteenKDE)
Colors = KMfunction.predict(FourteenKDE)

plt.clf()
plt.figure(figsize=(4, 4))
plt.scatter(FourteenKDE['Longitude'],FourteenKDE['Latitude'],s=60, c = Colors, cmap = 'tab10',alpha = .4, linewidths = 0)
plt.title('Milwaukee Crime Events: Fourteen [K = 6]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2014\Euclidean\Year\Milwaukee Crime Events Fourteen [K=6].png')
plt.show()

#******************************************************************************************************************************************************

KMfunction = KMeans(n_clusters = 7)
KMfunction.fit(FourteenKDE)
Colors = KMfunction.predict(FourteenKDE)

plt.clf()
plt.figure(figsize=(4, 4))
plt.scatter(FourteenKDE['Longitude'],FourteenKDE['Latitude'],s=60, c = Colors, cmap = 'tab10',alpha = .4, linewidths = 0)
plt.title('Milwaukee Crime Events: Fourteen [K = 7]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2014\Euclidean\Year\Milwaukee Crime Events Fourteen [K=7].png')
plt.show()

#******************************************************************************************************************************************************

```
