```python
SepNineMKE = SepNine [(SepNine['Longitude']>-88.080736) & (SepNine['Longitude']<-87.839722
)& (SepNine['Latitude']>42.91767) & (SepNine['Latitude']< 43.19712)]
# Create a boolean array setting boundaries to the map, Attribute these booleans to the columns of the original dataset

X = SepNineMKE['Longitude']
y = SepNineMKE['Latitude']


plt.clf()
plt.figure(figsize=(8, 8))
plt.scatter(X, y, s=60, alpha=.4, linewidths=0)
plt.title('Milwaukee Crime Events: SepNine [Cropped]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2009\Euclidean\Sep\Milwaukee Crime Events SepNine [Cropped].png')
plt.show() 

SepNineKDE = pd.DataFrame(SepNineMKE, columns = ['Longitude','Latitude'])

KMfunction = KMeans(n_clusters = 5)
KMfunction.fit(SepNineKDE)
Colors = KMfunction.predict(SepNineKDE)

plt.clf()
plt.figure(figsize=(4, 4))
plt.scatter(SepNineKDE['Longitude'],SepNineKDE['Latitude'],s=60, c = Colors, cmap = 'tab10',alpha = .4, linewidths = 0)
plt.title('Milwaukee Crime Events: SepNine [K = 5]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2009\Euclidean\Sep\Milwaukee Crime Events SepNine [K=5].png')
plt.show()

#******************************************************************************************************************************************************

KMfunction = KMeans(n_clusters = 4)
KMfunction.fit(SepNineKDE)
Colors = KMfunction.predict(SepNineKDE)

plt.clf()
plt.figure(figsize=(4, 4))
plt.scatter(SepNineKDE['Longitude'],SepNineKDE['Latitude'],s=60, c = Colors, cmap = 'tab10',alpha = .4, linewidths = 0)
plt.title('Milwaukee Crime Events: SepNine [K = 4]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2009\Euclidean\Sep\Milwaukee Crime Events SepNine [K=4].png')
plt.show()

#******************************************************************************************************************************************************

KMfunction = KMeans(n_clusters = 3)
KMfunction.fit(SepNineKDE)
Colors = KMfunction.predict(SepNineKDE)

plt.clf()
plt.figure(figsize=(4, 4))
plt.scatter(SepNineKDE['Longitude'],SepNineKDE['Latitude'],s=60, c = Colors, cmap = 'tab10',alpha = .4, linewidths = 0)
plt.title('Milwaukee Crime Events: SepNine [K = 3]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2009\Euclidean\Sep\Milwaukee Crime Events SepNine [K=3].png')
plt.show()

#******************************************************************************************************************************************************

KMfunction = KMeans(n_clusters = 6)
KMfunction.fit(SepNineKDE)
Colors = KMfunction.predict(SepNineKDE)

plt.clf()
plt.figure(figsize=(4, 4))
plt.scatter(SepNineKDE['Longitude'],SepNineKDE['Latitude'],s=60, c = Colors, cmap = 'tab10',alpha = .4, linewidths = 0)
plt.title('Milwaukee Crime Events: SepNine [K = 6]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2009\Euclidean\Sep\Milwaukee Crime Events SepNine [K=6].png')
plt.show()

#******************************************************************************************************************************************************

KMfunction = KMeans(n_clusters = 7)
KMfunction.fit(SepNineKDE)
Colors = KMfunction.predict(SepNineKDE)

plt.clf()
plt.figure(figsize=(4, 4))
plt.scatter(SepNineKDE['Longitude'],SepNineKDE['Latitude'],s=60, c = Colors, cmap = 'tab10',alpha = .4, linewidths = 0)
plt.title('Milwaukee Crime Events: SepNine [K = 7]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig(r'C:\Users\qnche\OneDrive\Desktop\Marquette REU Summer 2018\K means images\2009\Euclidean\Sep\Milwaukee Crime Events SepNine [K=7].png')
plt.show()

#******************************************************************************************************************************************************

```
